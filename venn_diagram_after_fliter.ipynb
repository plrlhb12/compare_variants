```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
#mpl.rcParams.update(mpl.rcParamsDefault)
%config InlineBackend.figure_format = 'retina'
```

```python
from matplotlib_venn import venn3, venn3_circles
from matplotlib_venn import venn2, venn2_circles
```

```python
color1 = "darkorange"
color2 = 'teal'
color3 = "blue"
color4 = "seagreen"
```

```python
# read 3 sets of text files
# these files were copied from biowulf /data/CARD/tprojects/data/temp/short/
# 4 fields of chr, pos, REF, ALT were merged into one field
a = pd.read_csv("UNHS_GT19-38445_bcfc1_cut_merage.vcf", sep="\t", header=None)
a.columns = ["A"]
a.head()
```

```python
b = pd.read_csv("Jax_KOLF2-ARID2-A2_bcfc1_cut_merage.vcf", sep="\t", header=None)
b.columns = ["B"]
b.head()
```

```python
c = pd.read_csv("long_psomagen_KOLF2.1_variants_filt30_merge.vcf", sep="\t", header=None)
c.columns = ["C"]
c.head()
```

```python
d = pd.read_csv("long_Jax_KOLF2C1_variants_filt30_merge.vcf",sep="\t", header=None)
d.columns = ["D"]
```

```python
_=venn3([set(a["A"]), set(b["B"]), set(c["C"])], 
        set_labels = ('Psomagen_KOLF2.1_short', 'Jax_KOLF2.1_short', 'Psomagen_KOLF2.1_long'),
       set_colors=(color1, color2, color3))
plt.title("Three sets of KOLF2.1 after filter")
plt.savefig("3_kolf2.1_filter.pdf")
plt.savefig("3_kolf2.1_filter.png")
```
